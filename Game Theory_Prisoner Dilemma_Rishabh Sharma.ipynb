```python
# import random package
import random 

#Defining the problem
print("""
    One of the most well-known examples of game theory is the \"Prisoner's Dilemma."
    Game theory is a branch of mathematics that helps to analyze and understand 
    social interactions in which individuals compete with each other.
    Let us play the game to see how this theory works.
""")
print("-"*80)

print("""
    You will get 2 options to start with:
    1. Play against the computer
    2. Let the computer play 5 rounds with two imaginary players
    
    In the first scenario you will get to choose a strategy either dominant or 
    co-operative and the computer will choose its own strategy to win the game.
    In the second scenario the two computer players will have their own strategies
    for each round based on the results of the previous rounds upto 5 rounds. 
    After the 5th round you will be able to see the results of each round and final 
    result based on the points collected in all 5 rounds.    
    """)

print("-"*80)

#Ask the user if they want to play with computer or not
user_play = input("""
    Do you want to play a Game Theory Situation?: yes/no """).lower()

# Game Matrix.
score = [[5, 3], [10, 0]]

# Save board points and results in a dictionary. R: remain silent, B: betray
dictionary_results = {"(R, R)": (score[0][0], score[0][0]),
                      "(R, B)": (score[0][1], score[1][0]),
                      "(B, R)": (score[1][0], score[0][1]),
                      "(B, B)": (score[1][1], score[1][1])}



# If user want to play
if "yes" in user_play:
    player_name = input("""
    Hi, and welcome to this game! You will play against me! Whats your name? """).capitalize()
    
    print(f"\n{player_name} We were caught! Do not confess! We didn't do anything!")
    
    print("""
    You are in small room at the police station. You can't communicate with anyone.
    One of the policeman approach to you and tells you:""")
    
    # Game Situation
    print(f"""
    Policeman: {player_name}, this is your situation, think well before answering.
    You can either stay silent or choose to snitch. If you choose to remain silent
    and your partner snitches you get 10 years imprisonment and your partner gets 
    1 year and vise versa. If you both snitch you both get 25 years. In case you 
    both remain silent and dont snitch you both get 3 years. Now its up to you 
    either sell out your partner and you have an opportunity to go back home in one 
    year or stay silent hoping that your partner also does the same
    
    \n
    {player_name} \ Computer\t\t Stay Silent \t\t\t  Snitch
    --------------\t --------------------------\t ---------------------------
    Stay Silent\t\t Both gets 3 years\t\t {player_name} gets 10 year, Python 1
    Snitch \t\t {player_name} gets 1 years, Python 10\t 25 years both
    """)
        
    # prompt the user to select what to do
    player_choice = input("""
    What would you do? Would you remain silent or cooperate with the police?
    Select:
    - 'R' to remain silent
    - 'B' to betray
    """).upper()
    
    # Computer (Python) choose randomly either to remaint silent or betray
    computer_choice = random.randint(0,1) # 0: remain silent; 1: betray
    
    # conditional structure with the outputs of each scenario
    if player_choice == "R" and computer_choice == 0:
        player_result = dictionary_results["(R, R)"][0]
        computer_result = dictionary_results["(R, R)"][1]
        print(f"""
    You've decided to remain silent, as well as me!
    You earned {player_result} points.
    I earned {computer_result}""")
    
    elif player_choice == "B" and computer_choice == 0:
        player_result = dictionary_results["(B, R)"][0]
        computer_result = dictionary_results["(B, R)"][1]
        print(f"""
    You've decided to snitch, I saved you!
    You earned {player_result} points.
    I earned {computer_result}""")

    elif player_choice == "R" and computer_choice == 1:
        player_result = dictionary_results["(R, B)"][0]
        computer_result = dictionary_results["(R, B)"][1]
        print(f"""
    I snitch you, you saved me!
    You earned {player_result} points.
    I earned {computer_result}""")

    elif player_choice == "B" and computer_choice == 1:
        player_result = dictionary_results["(B, B)"][0]
        computer_result = dictionary_results["(B, B)"][1]
        print(f"""
    We both snitched, we both go to jail
    You earned {player_result} points.
    I earned {computer_result}""")
    
    else:
        print("Someting went wrong!! please choose either R or B")
        
    #Declaring results
    if player_result > computer_result:
        print(f"""
    Congrats {player_name}, you won!""")

    elif player_result == computer_result:
        print(f"""
    Well we tie!""")

    elif player_result <= computer_result:
        print(f"""
    Good efforts but you lost!""")
        
    else:
        print("Error in evaluating results")


elif "no" in user_play:
    print("""
    So! you deicded not to play the game.
    No worries, it can be difficult a difficult choice to save yourself or your 
    friend. Now we will make the choice for you.
    """)
    
    print("-"*80)
    
    print("""
    Here the computer will be playing 5 rounds for you. Just select the player names.
    Once you select the player names we will be sharing the choices each player selected
    per round and the points they scored. 
    At the end we will share the total points each player scored in each round
    along with the final winner.
    """)
    
    player_1 = input("""
    Hi, please select the name of first player: """).capitalize()
    
    player_2 = input("""
    Hi, please select the name of second player: """).capitalize()
    
    print(f""" 
    {player_1} and {player_2} were caught!""")
    
    # Game Situation
    print(f"""
    Policeman: This is the situation. Think well before answering.
    You can either stay silent or choose to snitch. If you choose to remain silent
    and your partner snitches you get 10 years imprisonment and your partner gets 
    1 year and vise versa. If you both snitch you both get 25 years. In case you 
    both remain silent and dont snitch you both get 3 years. Now its up to you 
    either sell out your partner and you have an opportunity to go back home in one 
    year or stay silent 
    
    \n
    {player_1}\ {player_2}\t\t Stay Silent \t\t\t  Snitch
    --------------\t --------------------------\t ---------------------------
    Stay Silent\t\tBoth gets 3 years\t\t{player_1} gets 10 year, {player_2} 1
    Snitch \t\t {player_1} gets 1 years,{player_2} 10\t\t 25 years both
    """)
    
    total_rounds = 5
    
    # Round 1 - 
    # player 1 choose randomly 
    p1_r1_choice = random.randint(0,1) # 0: remain silent; 1: betray
    
    # player 2 uses a generous tit for tat strategy here
    p2_r1_choice = "B"
    
    # conditional structure with the outputs of each scenario
    if p2_r1_choice == "R" and p1_r1_choice == 0:
        p2_r1_result = dictionary_results["(R, R)"][0]
        p1_r1_result = dictionary_results["(R, R)"][1]

    elif p2_r1_choice == "B" and p1_r1_choice == 0:
        p2_r1_result = dictionary_results["(B, R)"][0]
        p1_r1_result = dictionary_results["(B, R)"][1]
       
    elif p2_r1_choice == "R" and p1_r1_choice == 1:
        p2_r1_result = dictionary_results["(R, B)"][0]
        p1_r1_result = dictionary_results["(R, B)"][1]
       
    elif p2_r1_choice == "B" and p1_r1_choice == 1:
        p2_r1_result = dictionary_results["(B, B)"][0]
        p1_r1_result = dictionary_results["(B, B)"][1]
        
    else:
        print("Someting went wrong!!")
        
    print("""
    There are 2 choices for each player
    To remain silent by cooperating or to betray your partner?
    Select:
     - 'R' or 0 to remain silent
     - 'B' or 1 to betray
        """)

    p1_choice = [p1_r1_choice] # list of p1 choices
    
    p2_choice = [p2_r1_choice] # list of p2 choices
     
    p1_points = [p1_r1_result] # List of p1 results
    
    p2_points = [p2_r1_result] # list of p2 results
    
    #number of rounds to be played
    remaining_rounds = 4
    
    while remaining_rounds >= 0:

        print(f"""
    Round: {total_rounds - remaining_rounds}
    {player_1} choice : {p1_choice[-1]}
    {player_2} choice : {p2_choice[-1]}
    {player_1} points : {p1_points[-1]}
    {player_2} points : {p2_points[-1]}     
        """)
        
        if p1_choice[-1] == 0:
            
            player2_choice = "R"
                    
        elif p1_choice[-1] == 1:
            
            player2_choice = "B"
            
    # player 1 choose randomly 
        player1_choice = random.randint(0,1) # 0: remain silent; 1: betray
    
    # conditional structure with the outputs of each scenario
        if player2_choice == "R" and player1_choice == 0:
            p2_result = dictionary_results["(R, R)"][0]
            p1_result = dictionary_results["(R, R)"][1]

        elif player2_choice == "B" and player1_choice == 0:
            p2_result = dictionary_results["(B, R)"][0]
            p1_result = dictionary_results["(B, R)"][1]

        elif player2_choice == "R" and player1_choice == 1:
            p2_result = dictionary_results["(R, B)"][0]
            p1_result = dictionary_results["(R, B)"][1]

        elif player2_choice == "B" and player1_choice == 1:
            p2_result = dictionary_results["(B, B)"][0]
            p1_result = dictionary_results["(B, B)"][1]

        else:
            print("Someting went wrong!!")

        p1_choice.append(player1_choice)

        p2_choice.append(player2_choice)

        p1_points.append(p1_result)

        p2_points.append(p2_result)    

        remaining_rounds -= 1
        
    p1_total = sum(p1_points) # calculating total points scored by player 1
    
    p2_total = sum(p2_points) # calculating total points scored by player 2
    
    print(f"""
    The strategy used by {player_1} is random where the player chooses their
    outcomes randomly and the total points earned with this strategy: {p1_total}

    The strategy used by {player_2} is suspecious tit for tat where the player 
    chooses starts by betraying and then strats imitating the decisions made by
    their partner in the previous round and the total points earned with this 
    strategy: {p2_total}    
    """)
    
    #Declaring results
    if p1_total > p2_total:
        
        print(f"""
    Congratulations {player_1}, you won!""")
        
    elif p1_total < p2_total:
        
        print(f"""
    Congratulations {player_2}, you won!""")
        
    elif p1_total == p2_total:
        
        print("""
    It's a tie!""")     
        
else:
    print("something went wrong please check your input")    
    
```
